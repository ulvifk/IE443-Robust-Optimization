```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.spatial.distance as distance
from scipy.stats import chi2
import math
```

```python

data = pd.read_csv("dataset3.csv")
data = data.append({"x1": 18, "x2": 18, "x3": 18,}, ignore_index=True)
data
```

```python
data.cov()
```

```python
x1 = data["x1"]
x2 = data["x2"]
x3 = data["x3"]

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

ax.scatter(x1,x2,x3)
fig.show()
```

```python
data.mean()
```

```python
md = []
for index, row in data.iterrows():
    md.append(distance.mahalanobis([row["x1"], row["x2"], row["x3"]], data.mean(), data.cov()))

data["Mahalanobis"] = md
data['p_value'] = 1 - chi2.cdf(data['Mahalanobis'], 2)
data
```

```python
max(data["Mahalanobis"])
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()

ax.hist(data["Mahalanobis"], bins = int(math.sqrt(data.shape[0])))
```

```python
cutoff = data["Mahalanobis"].mean() + 3 * data["Mahalanobis"].std()
cutoff
```

```python
outlierIndexes = np.where(data["Mahalanobis"] > cutoff)
outlierIndexes
```

```python
ax.scatter(data["x1"].loc[outlierIndexes], data["x2"].loc[outlierIndexes], data["x3"].loc[outlierIndexes])
fig.show()
```
